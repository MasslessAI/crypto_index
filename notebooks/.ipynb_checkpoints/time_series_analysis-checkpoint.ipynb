```python
import sys
sys.path.append('../')

import cqt
import cqt.datagen as dg
import cqt.model.asset_model as md
import cqt.model.asset_model_component_spot as comp_spot
import pandas as pd
import numpy as np

from datetime import datetime
from datetime import timedelta
import matplotlib.pyplot as plt
from matplotlib import pyplot

from statsmodels.tsa import arima_model as arma
from sklearn.metrics import mean_squared_error
```

```python
df_btc_1d = pd.read_csv('..\data\csv\coinapi_ohlcv_GEMINI_SPOT_BTC_USD_1DAY_data.csv')
df_eth_1d = pd.read_csv('..\data\csv\coinapi_ohlcv_GEMINI_SPOT_ETH_USD_1DAY_data.csv')
df_btc_1h = pd.read_csv('..\data\csv\coinapi_ohlcv_GEMINI_SPOT_BTC_USD_1HRS_data.csv')
df_eth_1h = pd.read_csv('..\data\csv\coinapi_ohlcv_GEMINI_SPOT_ETH_USD_1HRS_data.csv')
```

```python
df_btc_1d.head()
```

```python
model_config = {'asset_type': 'spot'}
model_btc_1d = comp_spot.AssetModelComponentSpot('btc', df_btc_1d, model_config)
model_eth_1d = comp_spot.AssetModelComponentSpot('eth', df_eth_1d, model_config)
model_btc_1d.plot_price()
model_eth_1d.plot_price()
asset_model_1d = md.AssetModel([])
asset_model_1d.insert_component(model_btc_1d)
asset_model_1d.insert_component(model_eth_1d)
price_btc_eth_1d = asset_model_1d.concat_prices('btc','eth')
logret_btc_eth_1d = asset_model_1d.concat_log_return('btc','eth', '1d')

model_btc_1h = md.AssetModelComponent('btc', df_btc_1h, model_config)
model_eth_1h = md.AssetModelComponent('eth', df_eth_1h, model_config)
model_btc_1h.plot_price()
model_eth_1h.plot_price()
asset_model_1h = md.AssetModel([])
asset_model_1h.insert_component(model_btc_1h)
asset_model_1h.insert_component(model_eth_1h)
price_btc_eth_1h = asset_model_1h.concat_prices('btc','eth')
logret_btc_eth_1h = asset_model_1h.concat_log_return('btc','eth', '1d')
```

```python
#ADF stationarity test, check for time-dependent structure

from statsmodels.tsa.stattools import adfuller
af_test = adfuller(logret_btc_eth_1d['btc'])
print('ADF Statistic: %f' % af_test[0])
print('p-value: %f' % af_test[1])
print('Critical Values:')
for key, value in af_test[4].items():
    print('\t%s: %.3f' % (key, value))
    
af_test = adfuller(logret_btc_eth_1d['eth'])
print('ADF Statistic: %f' % af_test[0])
print('p-value: %f' % af_test[1])
print('Critical Values:')
for key, value in af_test[4].items():
    print('\t%s: %.3f' % (key, value))
    
af_test = adfuller(logret_btc_eth_1h['btc'])
print('ADF Statistic: %f' % af_test[0])
print('p-value: %f' % af_test[1])
print('Critical Values:')
for key, value in af_test[4].items():
    print('\t%s: %.3f' % (key, value))
    
af_test = adfuller(logret_btc_eth_1h['eth'])
print('ADF Statistic: %f' % af_test[0])
print('p-value: %f' % af_test[1])
print('Critical Values:')
for key, value in af_test[4].items():
    print('\t%s: %.3f' % (key, value))       
```

```python
# Ljung-Box test for autocorrelation

# High freq data usually behaves like autocorrelated.

from statsmodels.stats.diagnostic import acorr_ljungbox
acorr_test = acorr_ljungbox(logret_btc_eth_1d['btc'])
print(acorr_test[1])
acorr_test = acorr_ljungbox(logret_btc_eth_1h['btc'])
print(acorr_test[1])
acorr_test = acorr_ljungbox(logret_btc_eth_1d['eth'])
print(acorr_test[1])
acorr_test = acorr_ljungbox(logret_btc_eth_1h['eth'])
print(acorr_test[1])
```

```python
#Test ARIMA fit, time dependency

X = logret_btc_eth_1d['btc'].rolling(window=3).mean().dropna()
X = X * 100
#BTC Daily
arma_btc_1d = arma.ARMA(X, order=(5,1))
arma_btc_1d_fit = arma_btc_1d.fit(disp=0)
print(arma_btc_1d_fit.summary())
# plot residual errors
residuals = pd.DataFrame(arma_btc_1d_fit.resid)
residuals.plot()
pyplot.show()
residuals.plot(kind='kde')
pyplot.show()
print(residuals.describe())
```

```python
aic_table = pd.DataFrame()
aic_table.column = ['0', '1']
for i in range(12):
    for j in range(2):
        model = arma.ARMA(X, order=(i,j))
        model_fit = model.fit(disp=0)
        aic_table.set_value(i, j, model_fit.aic)
        
aic_table.plot()
plt.show()
```

```python
#ARIMA predicts

btc = logret_btc_eth_1d['btc'].rolling(window=30).mean().dropna()
btc = btc * 100

train_size = int(len(btc) * 0.66)
train = btc[0:train_size]
test = btc[train_size:len(btc)]
history = [x for x in train]
predictions_1 = list()
predictions_n = list()

for t in range(len(test)):
    try:
        model = arma.ARMA(history, order=(7, 0))
        model_fit = model.fit(disp=0)
        output = model_fit.forecast()
        yhat = output[0]
        predictions_1.append(yhat)
        obs = test[t]
        history.append(obs)
        print('predicted=%f, expected=%f' % (yhat, obs))
    except:
        pass
#         predictions.append(0.0)
error = mean_squared_error(test.values, predictions_1)
print('Test MSE: %.3f' % error)

model = arma.ARMA(history, order=(7, 0))
model_fit = model.fit(disp=0)
output = model_fit.forecast(t)
predictions_n = output[0]

# #         predictions.append(0.0)
# error = mean_squared_error(test.values, predictions_n)
# print('Test MSE: %.3f' % error)


pyplot.plot(test.values, color='black')
pyplot.plot(predictions_1, color='red')
pyplot.plot(predictions_n, color='green')
pyplot.show()
```
